```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import animation, rc
from IPython.display import HTML
import scipy
```

Bessel Functions of the first and the second kind

```python
fig = plt.figure(figsize=(16, 9), dpi=1080/18)
ax = fig.add_subplot(111, autoscale_on=False, xlim=(0, 15), ylim=(-1, 1))
ax.grid()
lines = []

a_text = ax.text(0.02, 0.80, '', transform=ax.transAxes, fontsize=30)

def sincfunc(x, a):
    return 1 / (np.pi) / x * np.sin(x / a) 

def animate(v):
    if len(lines):
        lines.pop() 
    x = np.linspace(0, 15, 1000)
    y = scipy.special.jv(v, x)
    new_line, = ax.plot(x, y, '-', lw=2)
    lines.append(new_line)
    a_text.set_text('Bessel J Order = {}'.format(v))
    lines.append(a_text)
    return lines

anim = animation.FuncAnimation(
    fig, 
    animate, 
    frames=range(10), 
    interval=1000, 
    # blit=True,
    repeat=False,
)
plt.close(fig)
HTML(anim.to_html5_video())
```
